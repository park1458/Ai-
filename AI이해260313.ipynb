```
# 사칙연산(+, -, *, /)을 하는 아주 간단한 프로그램
# (변수, 연산자, 입력문(input), 출력문(print) 사용)

# 1) 사용자에게 값을 입력받기
# input()은 "문자"로 입력을 받기 때문에, int()로 정수로 바꿔서 변수에 저장합니다.
a = int(input("첫 번째 숫자 a를 입력하세요: "))
b = int(input("두 번째 숫자 b를 입력하세요: "))

# 2) 사칙연산 결과를 각각 변수에 저장
add_result = a + b      # 더하기
sub_result = a - b      # 빼기
mul_result = a * b      # 곱하기
div_result = a / b      # 나누기(/는 실수 나눗셈이라 소수가 나올 수 있음)

# 3) 결과 출력
print("a =", a)
print("b =", b)

print("a + b =", add_result)
print("a - b =", sub_result)
print("a * b =", mul_result)
print("a / b =", div_result)
```
