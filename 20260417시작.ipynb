```
# 카페 메뉴(원)
menu = {
    "아메리카노": 4500,
    "카페라떼": 5200,
    "바닐라라떼": 5700,
    "카푸치노": 5200,
    "카라멜마키아토": 5900,
    "콜드브루": 5500,
    "카페모카": 6000,
    "자몽에이드": 5800,
    "레몬에이드": 5800,
    "초코라떼": 5500,
}

order = {}  # 예: {"아메리카노": 2, "카페라떼": 1}

print("=== 카페 주문 ===")
print("메뉴:", ", ".join(menu.keys()))
print("입력 예) 아메리카노 2")
print("취소: '취소 메뉴명'  또는  '전체취소'")
print("결제: '결제'  |  종료: '종료'")

while True:
    user = input("\n주문 입력: ").strip()

    if user == "종료":
        print("프로그램을 종료합니다.")
        break

    if user == "결제":
        if not order:
            print("주문 내역이 없습니다.")
            continue

        total = 0
        print("\n===== 영수증 =====")
        for name, qty in order.items():
            price = menu[name]
            line = price * qty
            total += line
            print(f"{name: <10} x {qty: <2} = {line:,}원")
        print("------------------")
        print(f"총액: {total:,}원")
        print("==================")
        break

    if user == "전체취소":
        order.clear()
        print("전체 주문을 취소했습니다.")
        continue

    # "취소 메뉴명" 처리
    if user.startswith("취소 "):
        name = user[3:].strip()
        if name in order:
            del order[name]
            print(f"'{name}' 주문을 취소했습니다.")
        else:
            print(f"'{name}'는 주문 내역에 없습니다.")
        continue

    # 일반 주문: "메뉴명 수량"
    parts = user.split()
    if len(parts) != 2:
        print("형식이 올바르지 않습니다. 예) 아메리카노 2")
        continue

    name, qty_str = parts[0], parts[1]

    if name not in menu:
        print("없는 메뉴입니다. 메뉴판을 확인해주세요.")
        continue

    if not qty_str.isdigit():
        print("수량은 숫자로 입력해주세요.")
        continue

    qty = int(qty_str)
    if qty <= 0:
        print("수량은 1 이상이어야 합니다.")
        continue

    order[name] = order.get(name, 0) + qty
    print(f"추가됨: {name} {qty}잔 (현재 {order[name]}잔)")
```
