```
# 사칙연산(+, -, *, /)을 하는 아주 간단한 프로그램
# (변수, 연산자, 입력문(input), 출력문(print) 사용)

# 1) 사용자에게 값을 입력받기
# input()은 "문자"로 입력을 받기 때문에, int()로 정수로 바꿔서 변수에 저장합니다.
a = int(input("첫 번째 숫자 a를 입력하세요: "))
b = int(input("두 번째 숫자 b를 입력하세요: "))

# 2) 사칙연산 결과를 각각 변수에 저장
add_result = a + b      # 더하기
sub_result = a - b      # 빼기
mul_result = a * b      # 곱하기
div_result = a / b      # 나누기(/는 실수 나눗셈이라 소수가 나올 수 있음)

# 3) 결과 출력
print("a =", a)
print("b =", b)

print("a + b =", add_result)
print("a - b =", sub_result)
print("a * b =", mul_result)
print("a / b =", div_result)
```

```
def is_leap_year(year: int) -> bool:
    # 윤년 규칙:
    # 1) 4로 나누어떨어지면 윤년 후보
    # 2) 그 중 100으로 나누어떨어지면 평년
    # 3) 단, 400으로 나누어떨어지면 다시 윤년
    return (year % 4 == 0) and ((year % 100 != 0) or (year % 400 == 0))


# 사용 예시
y = int(input("연도를 입력하세요: "))
print("윤년입니다." if is_leap_year(y) else "윤년이 아닙니다.")
```
