# 새 섹션

```
import pandas as pd
```

```
import os

# 파일이 있는지 확인하고 로드합니다.
file_path = 'Iris.csv'

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print("데이터 로드 완료!")
    display(df.head())
else:
    print(f"에러: '{file_path}' 파일을 찾을 수 없습니다. 파일이 노트북과 같은 폴더에 있는지 확인해주세요.")
```

```
sepal_length_column = 'SepalLengthCm'
average_sepal_length = df[sepal_length_column].mean()
print(f"평균 꽃받침 길이: {average_sepal_length:.2f}")
```

```
.petal_width_column = 'PetalWidthCm'
maximum_petal_width = df[petal_width_column].max()
print(f"최대 꽃잎 너비: {maximum_petal_width:.2f}")
```

```
sepal_length_gt_2 = df[df['SepalLengthCm'] > 2]
count_sepal_length_gt_2 = len(sepal_length_gt_2)
print(f"꽃받침 길이가 2보다 큰 데이터의 개수: {count_sepal_length_gt_2}")
```

```
ls = df.groupby('Species')['SepalLengthCm'].mean().idxmax()
print(f"꽃받침 길이가 가장 긴 종류: {ls}")
```

```
wp = df.groupby('Species')['PetalWidthCm'].mean().idxmax()
print(f"꽃잎이 가장 넓은 종류: {wp}")
```

```
df['sa'] = df['SepalLengthCm'] * df['SepalWidthCm']
lsa = df.groupby('Species')['sa'].mean().idxmax()
print(f"꽃받침 면적이 가장 큰 종류: {lsa}")
```

```
df['pa'] = df['PetalLengthCm'] * df['PetalWidthCm']
lpa = df.groupby('Species')['pa'].mean().idxmax()
print(f"꽃잎 면적이 가장 큰 종류: {lpa}")
```

```
fdf = df[(df['SepalWidthCm'] >= 3.0) & (df['PetalWidthCm'] >= 1.5)]
print("필터링된 데이터:")
display(fdf.head())
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.histplot(df['PetalLengthCm'], kde=True, bins=20)
plt.title('꽃잎 길이 (PetalLengthCm) 분포')
plt.xlabel('꽃잎 길이 (cm)')
plt.ylabel('빈도수')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', style='Species', s=100)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Species', y='SepalLengthCm', hue='Species', palette='viridis', legend=False)
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length (cm)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species', style='Species', s=100)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```
correlation = df['PetalLengthCm'].corr(df['PetalWidthCm'])
print(f"꽃잎 길이와 너비 사이의 상관계수: {correlation:.4f}")

if correlation > 0.7:
    print("상관계수가 매우 높으므로, 꽃잎의 길이가 길수록 너비가 넓어진다고 할 수 있습니다.")
elif correlation > 0.3:
    print("어느 정도 양의 상관관계가 있습니다.")
else:
    print("상관관계가 뚜렷하지 않습니다.")
```

```
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Species', y='SepalLengthCm', estimator='mean', errorbar=None, hue='Species', palette='muted')
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
