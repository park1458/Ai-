```
even_sum = 0
odd_sum = 0

for i in range(1, 101):
    if i % 2 == 0:      # 짝수
        even_sum += i
    else:               # 홀수
        odd_sum += i

print("짝수합:", even_sum)
print("홀수합:", odd_sum)
```
